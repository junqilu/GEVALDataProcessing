# Background data

## Read in data

```python
background_data_folder_path = 'input\Background_data'

csv_files = [file for file in os.listdir(background_data_folder_path) if
             file.endswith('.csv')]
csv_files
```

```python
background_data_df_list = []

for file in csv_files:
    file_path = os.path.join(background_data_folder_path, file)
    df = pd.read_csv(file_path)
    background_data_df_list.append(df)

merged_df = pd.concat(background_data_df_list, ignore_index=True)
merged_df
```

## Clean up

```python
#Drop the index col generated by Fiji
merged_df.drop(columns=[' '], inplace=True)
merged_df.reset_index(inplace=True, drop=True)
merged_df
```

```python
merged_df[['file_name']] = merged_df['Label'].str.extract('^(.*.ims):')
merged_df[['channel']] = merged_df['Label'].str.extract('(\d{3})$')
merged_df.drop(columns=['Label'], inplace=True)

cols = list(merged_df.columns)
cols = ['file_name', 'channel'] + [col for col in cols if col not in
                               ['file_name', 'channel']] #Move the col
# 'file_name', 'channel' to be the 1st 2 cols
merged_df = merged_df[cols]

merged_df
```

## Pivot

```python
agg_dict = {}
for col in merged_df.columns:
    if col in [
        'Mean' #I only care about the parameters related to the intensity
        # and all the ones related to the shape are irrelevant
    ]:
        agg_dict[col]=['mean', 'median'] #I don't care about the mode for
        # the background data, so the mode is excluded here
agg_dict
```

```python
# Group by file_name and channel, then aggregate using mean, median, and mode
grouped_df = merged_df.groupby(['file_name', 'channel']).agg(agg_dict)

# Flatten the hierarchical columns
grouped_df.columns = ['_'.join(col).strip() for col in grouped_df.columns.values]

# Reset index to make file_name and channel regular columns
grouped_df.reset_index(inplace=True)

# Pivot the DataFrame so each file_name has only one row of data
background_result_df = grouped_df.pivot(index='file_name', columns='channel')

# Flatten the hierarchical columns after pivoting
background_result_df.columns = ['{}_{}'.format(col[0], col[1]) for col in background_result_df.columns]

# Reset index to make file_name a regular column
background_result_df.reset_index(inplace=True)

background_result_df
```

# Histogram data
For each histogram data, you should normalize the count by the sum of all
the counts in that histogram data since each cell_traced has a different
area (their total counts will be different)

```python
histogram_data_folder_path = 'input\Histogram_data'
```

## 405 channel

### Read in data

```python
histogram_405_csv_files = [file for file in os.listdir
(histogram_data_folder_path) if
             file.endswith('.csv') and '405' in file]

histogram_405_csv_files
```

```python
# Read and merge CSV files
dfs = []
all_bin_start_list = []

# Step 1: Read each CSV file into a DataFrame and add a file_name column
for file in histogram_405_csv_files:
    df = pd.read_csv(os.path.join(histogram_data_folder_path, file))

    #Normalize the count by their total count. This should be helpful since
    # the cell has different area and the area info is already captured by
    # the shape_data
    sum_counts = df['count'].sum()
    df['normalized_count'] = df['count'] / sum_counts

    df['file_original_name'] = file

    all_bin_start_list += df['bin_start'].to_list()

    dfs.append(df)

all_bin_start_list_sorted = sorted(all_bin_start_list)
bin_start_unique_set = set(all_bin_start_list_sorted)

# Step 2: Create an empty DataFrame with the desired columns
merged_df = pd.DataFrame(columns=sorted(bin_start_unique_set))


# Step 3: Iterate over each DataFrame and append normalized counts to merged_df
for df in dfs:
    # Create a Series with normalized counts as values and bin_start as the index
    counts_series = pd.Series(df['normalized_count'].values, index=df['bin_start'])

    # Append the Series to the merged DataFrame, using file name as the index
    merged_df.loc[df['file_original_name'].iloc[0]] = counts_series

# Reset the index to make file_name a regular column
merged_df.reset_index(inplace=True)
merged_df.rename(columns={'index': 'file_original_name'}, inplace=True)
merged_df.fillna(0, inplace=True) #Fill up all the NaN as 0 in the merged_df
merged_df
```

### Clean up

```python
merged_df[['file_name']] = merged_df['file_original_name'].str.extract('\d{3}_(.*).csv$')
merged_df['file_name'] = merged_df['file_name'] + '.ims' #Recover the file
# name
merged_df.drop(columns=['file_original_name'], inplace=True)

cols = list(merged_df.columns)
cols = ['file_name'] + [col for col in cols if col not in
                               ['file_name']] #Move the col
# 'file_name', 'channel' to be the 1st 2 cols
merged_df = merged_df[cols]

histogram_405_data = merged_df.copy()
histogram_405_data
```

## heatmap

### Read in data

```python
histogram_heatmap_csv_files = [file for file in os.listdir
(histogram_data_folder_path) if
             file.endswith('.csv') and 'heatmap' in file]

histogram_heatmap_csv_files
```

```python
# Read and merge CSV files
dfs = []
all_bin_start_list = []

# Step 1: Read each CSV file into a DataFrame and add a file_name column
for file in histogram_heatmap_csv_files:
    df = pd.read_csv(os.path.join(histogram_data_folder_path, file))

    grouped_df = df.groupby('bin_start')['count'].sum().reset_index() #This
    # is necessary because some of the csv have duplicated values in
    # bin_start col and that will lead to "cannot reindex on an axis with
    # duplicated labels" error on the step of "merged_df.loc[df['file_original_name'].iloc[0]] = counts_series"

    #Normalize the count by their total count. This should be helpful since
    # the cell has different area and the area info is already captured by
    # the shape_data
    sum_counts = grouped_df['count'].sum()
    grouped_df['normalized_count'] = grouped_df['count'] / sum_counts

    grouped_df['file_original_name'] = file

    all_bin_start_list += grouped_df['bin_start'].to_list()

    dfs.append(grouped_df)

all_bin_start_list_sorted = sorted(all_bin_start_list)
bin_start_unique_set = set(all_bin_start_list_sorted)

# Step 2: Create an empty DataFrame with the desired columns
merged_df = pd.DataFrame(columns=sorted(bin_start_unique_set))


# Step 3: Iterate over each DataFrame and append normalized counts to merged_df
for df in dfs:
    # Create a Series with normalized counts as values and bin_start as the index
    counts_series = pd.Series(df['normalized_count'].values, index=df['bin_start'])

    try:
        merged_df.loc[df['file_original_name'].iloc[0]] = counts_series
    except:
        print(df['file_original_name'].iloc[0])



# Reset the index to make file_name a regular column
merged_df.reset_index(inplace=True)
merged_df.rename(columns={'index': 'file_original_name'}, inplace=True)
merged_df.fillna(0, inplace=True) #Fill up all the NaN as 0 in the merged_df
merged_df
```

### Clean up

```python
merged_df[['file_name']] = merged_df['file_original_name'].str.extract('^histogram_data_for_heatmap_(.*).csv$')
merged_df['file_name'] = merged_df['file_name'] + '.ims' #Recover the file
# name
merged_df.drop(columns=['file_original_name'], inplace=True)

cols = list(merged_df.columns)
cols = ['file_name'] + [col for col in cols if col not in
                               ['file_name']] #Move the col
# 'file_name', 'channel' to be the 1st 2 cols
merged_df = merged_df[cols]

histogram_heatmap_data = merged_df.copy()
histogram_heatmap_data
```
